```python
import matplotlib.pyplot as plt
import pandas as pd


data1 = pd.read_csv("Поступления от клиентов - 1.csv")
data2 = pd.read_csv("Поступления от клиентов - 2.csv")
data3 = pd.read_csv("Поступления от клиентов - 3.csv")
data4 = pd.read_csv("Поступления от клиентов - 4.csv")
data5 = pd.read_csv("Поступления от клиентов - 5.csv")
data6 = pd.read_csv("Поступления от клиентов - 6.csv")
```

```python
data1
```

```python
def kart_shuharta(data):
  data['Дата'] = pd.to_datetime(data['Дата'])
  data['Сумма'] = data['Сумма'].astype(str).str.replace(' ', '')
  data['Сумма'] = data['Сумма'].str.replace(',', '.')
  data['Сумма'] = pd.to_numeric(data['Сумма'], errors='coerce')

  mean = data['Сумма'].mean()
  std_dev = data['Сумма'].std()
  n = len(data)

  UCL = mean + 3 * (std_dev)
  LCL = mean - 3 * (std_dev)
  plt.figure(figsize=(12, 6))
  plt.plot(data['Дата'], data['Сумма'], marker='o', linestyle='-', color='b', label='Сумма')
  plt.axhline(y=mean, color='r', linestyle='--', label='Среднее')
  plt.axhline(y=UCL, color='g', linestyle='--', label='UCL')
  plt.axhline(y=LCL, color='orange', linestyle='--', label='LCL')
  plt.title(f'Карта индивидуальных значений')
  plt.xlabel('Дата')
  plt.ylabel('Сумма')
  plt.xticks(rotation=45)
  plt.legend()
  plt.tight_layout()
  plt.show()


  data['Размах'] = data['Сумма'].diff().abs()
  R = data['Размах'].mean()
  UCL_range = 3.267 * R
  LCL_range = 0 * R
  plt.figure(figsize=(12, 6))
  plt.plot(data['Дата'], data['Размах'], marker='o', color='r', linestyle='--', label='суммы')
  plt.axhline(y=UCL_range, color='g', linestyle='--', label='UCL')
  plt.axhline(y=LCL_range, color='orange', linestyle='--', label='LCL')
  plt.axhline(y=R, color='b', linestyle='-', label='средний размах')
  plt.title('Карта размахов')
  plt.xlabel('дата')
  plt.ylabel('размах')
  plt.xticks(rotation=45)
  plt.legend()
  plt.tight_layout()
  plt.show()
```

```python
kart_shuharta(data1)
```
